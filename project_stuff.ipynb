```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r'C:\Users\jawyp\OneDrive\Desktop\precinct-data.csv')
```

```python
df = df[["GEOID20", 'Total_2020_Total', 'Total_2020_Pres', 'Dem_2020_Pres', 'Rep_2020_Pres', 'White_2020_Total', 'Black_2020_Total', 'Hispanic_2020_Total', 'Asian_2020_Total', 'Native_2020_Total', 'Pacific_2020_Total']]
```

```python
df['Dem_Percentage'] = df['Dem_2020_Pres'] / df['Total_2020_Pres']
df['Rep_Percentage'] = df['Rep_2020_Pres'] / df['Total_2020_Pres']
df['White_Percentage'] = df['White_2020_Total'] / df['Total_2020_Total']
df['Black_Percentage'] = df['Black_2020_Total'] / df['Total_2020_Total']
df['Hispanic_Percentage'] = df['Hispanic_2020_Total'] / df['Total_2020_Total']
df['Asian_Percentage'] = df['Asian_2020_Total'] / df['Total_2020_Total']
df['Native_Percentage'] = df['Native_2020_Total'] / df['Total_2020_Total']
df['Pacific_Percentage'] = df['Pacific_2020_Total'] / df['Total_2020_Total']
df['Nonwhite_Percentage'] = 1 - (df['White_2020_Total'] / df['Total_2020_Total'])
```

```python
df.fillna(0, inplace = True)
```

```python
df = df[df.Total_2020_Total > 100]
```

```python
X = df[['White_Percentage', 'Black_Percentage', 'Hispanic_Percentage', 'Asian_Percentage', 'Native_Percentage', 'Pacific_Percentage']]
y = df['Dem_Percentage']
model = linear_model.LinearRegression()
model.fit(X, y)
y_hat = model.predict(X)
print(f'R^2 for all groups is {model.score(X, y):.3}')
```
